## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
````

````python
# make a copy of the dataset
hbdf=df.copy()
````

````python
# drop 31994 duplicates rows
hbdf.drop_duplicates(inplace=True)
````

````python
# check shape of the dataset means how many rows and columns
hbdf.shape
````

````python
# check columns who have null values
hbdf.isnull().sum().sort_values(ascending=False)
````

````python
# drop company column because it has so many null values
hbdf.drop("company", axis='columns',inplace=True)
````

````python
# now again check columns for null values
hbdf.isnull().sum().sort_values(ascending=False)
````

````python
# agent column has 12193 null values
# fill null values with 'unknown'
hbdf["agent"].fillna("unknown", inplace = True)
````

````python
# country column has 452 null values
# fill null values with 'others'
hbdf["country"].fillna("others", inplace = True)
````

````python
# children column has only 4 null values
# fill null values with 0
hbdf["children"].fillna(0, inplace = True)
````

````python
# check for null values each column
# There are no columns have null values
hbdf.isnull().sum()
````

### **Now we can create some new column by the help of  previous columns**

````python
# create a new column by adding "adults" ,"children" and "babies" columns
hbdf["total_member"]=hbdf["adults"]+hbdf["children"]+hbdf["babies"]
````

````python
# create a new  column 'total_stay' by adding "stays_in_weekend_nights","stays_in_week_nights" columns
hbdf["total_stay"]=hbdf["stays_in_weekend_nights"]+hbdf["stays_in_week_nights"]
````

````python
# now check new shape of our dataset
hbdf.shape
````

````python
# check "total_member" column where values are 0
hbdf[hbdf["total_member"]==0].shape
````

### we found 166 rows in "total_member" where values are 0

````python
# drop those columns where 'total_member' are 0
hbdf=hbdf.drop(hbdf[hbdf['total_member'] ==0].index)
````

````python
# # check "total_member" column where values are 0
hbdf[hbdf['total_member'] ==0]
````
